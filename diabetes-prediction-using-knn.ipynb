We notice that the minimum value of certain fields is zero. The zero value for columns such as Glucose, etc does not make sense. This may be due to missing values

The following fields have an invalid zero value

Glucose,
BloodPressure,
SkinThickness,
Insulin,
BMI.

We can handle the invalid zero values by replacing them with NaN. This step will help count them easily. We need to replace the zeros with suitable values.

We copy the dataframe to a new dataframe and then replace the zero values in the above given 5 variables with NaN.

```python
#copy data to a new dataframe
data_copy = data.copy(deep=True)
```

```python
data_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = data_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NAN)
```

```python
data_copy.isnull().sum()
```

```python
data_copy
```

```python
plot = data.hist(figsize =(20,20))
```

Now we replace the NAN values with the mean value of the variable.

For each of the column, the NAN value is replaced with mean() value

```python
data_copy['Glucose'].fillna(data_copy['Glucose'].mean(),inplace=True)
data_copy['BloodPressure'].fillna(data_copy['BloodPressure'].mean(),inplace=True)
data_copy['SkinThickness'].fillna(data_copy['SkinThickness'].median(),inplace=True)
data_copy['Insulin'].fillna(data_copy['Insulin'].median(),inplace=True)
data_copy['BMI'].fillna(data_copy['BMI'].mean(),inplace=True)
```

```python
#Now we plot the histogram of updated data and analyze the changes due to the previous step

data_copy.hist(figsize=(20,20))
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
y = data_copy.Outcome
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=1/3,random_state=42,stratify=y)
```

# Data Modeling

```python

#Use KNN Algo
#We have used a for loop to fit the data on KNN model with K=1,......,15 and store the scores
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#to store the result

testing_score =[]
training_score=[]
```

```python
for i in range(1,15):
    knn = KNeighborsClassifier(i)
    knn.fit(X_train,y_train)
    
    
    training_score.append(knn.score(X_train,y_train))
    testing_score.append(knn.score(X_test,y_test))
    
```

```python
#Check out the value of K for which the training accuracy was the highest

max_training_score = max(training_score)

train_scores_ind = [i for i, v in enumerate(training_score) if v == max_training_score]
print('Max training score {} % and k = {}'.format(max_training_score*100,list(map(lambda x: x+1, train_scores_ind))))
```

```python
#Check out the value of K for which the testing accuracy was the highest

max_testing_score = max(testing_score)
test_scores_ind = [i for i, v in enumerate(testing_score) if v == max_testing_score]
print('Max testing score {} % and k = {}'.format(max_testing_score*100,list(map(lambda x: x+1, test_scores_ind))))
```

```python
# Hence for the given data set, we can select the KNN with K=11
```

```python

plt.plot(training_score,label='Training_score')
plt.plot(testing_score,label='Testing_score')
leg = plt.legend(loc='upper center')
plt.show()
```

```python
knn = KNeighborsClassifier(11)
```

```python
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
#prediction using test data
y_pred = knn.predict(X_test)
```

```python
#calculate model accuracy by comparing (y_test,y_pred)

from sklearn.metrics import accuracy_score

acc_knn = round(accuracy_score(y_test,y_pred) *100,2)
print('Accuracy of KNeighborsClassifier : ',acc_knn)
```
